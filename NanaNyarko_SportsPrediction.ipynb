```python
df_male_players = pd.read_csv("male_players (legacy).csv")
```

```python
df_male_players
```

```python
df_categorical = df_male_players.select_dtypes(include = object)
df_categorical
```

```python
df_quantitative = df_male_players.select_dtypes(include = np.number)
df_quantitative
```

```python
df_male_players.drop(columns=df_categorical,axis=1, inplace=True)
```

```python
df_male_players
```

```python
l = []
l_less = []

# Threshold for missing values (40% of total rows)
threshold = 0.4 * df_male_players.shape[0]

# Iterate over each column in the DataFrame
for column in df_male_players.columns:
    # Check if the number of missing values in the column is less than the threshold
    if df_male_players[column].isnull().sum() < threshold:
        l.append(column)  # Add column to list if it meets the criterion
    else:
        l_less.append(column)  # Add to another list if it does not meet the criterion

# Filter the DataFrame to include only the columns in the list 'l'
df_male_players = df_male_players[l]
df_male_players
```

```python
df_male_players.info()
```

```python
df_male_players = df_male_players.drop(columns=[
    'player_id',
    'fifa_version',
    'fifa_update',
    'league_id',
    'league_level',
    'club_team_id',
    'club_jersey_number',
    'club_contract_valid_until_year',
    'nationality_id',
    'international_reputation'
])
```

```python
df_male_players
```

```python
correlation = df_male_players.corr().iloc[0]

feature_correlation = []
threshold = 0.4

for i, x in correlation.items():
    if x > threshold:
        feature_correlation.append(i)
```

```python
feature_correlation
```

```python
df_male_players_new = df_male_players[feature_correlation]
```

```python
df_male_players_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.pipeline import Pipeline
```

```python
imputer = IterativeImputer(max_iter=10, random_state = 0)
```

```python
quantitative_pipeline = Pipeline([
    ('quant', imputer)
])
```

```python
df = pd.DataFrame(np.round(quantitative_pipeline.fit_transform(df_male_players_new)), columns = df_male_players_new.columns)
df
```

```python
y= df['overall']
x= df
x.drop("overall", inplace = True, axis=1)
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x= scaler.fit_transform(x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_test
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
l = LinearRegression()
```

```python
l.fit(x_train, y_train)
```

```python
predict_y = l.predict(x_test)
```

```python
print(f"Metrics for {l.__class__.__name__}:")
```

```python
print(f"The mean absolute error is {mean_absolute_error(predict_y, y_test)}\nThe mean squared error is {mean_squared_error(predict_y, y_test)}\nThe root mean squared error is {np.sqrt(mean_squared_error(predict_y, y_test))} \nThe R2 score is {r2_score(predict_y, y_test)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
regressor.fit(x_train, y_train)
```

```python
predict_y = regressor.predict(x_test)
```

```python
print(f"Metrics for {regressor.__class__.__name__}:")
print(f"The mean absolute error is {mean_absolute_error(predict_y, y_test)}\nThe mean squared error is {mean_squared_error(predict_y, y_test)}\nThe root mean squared error is {np.sqrt(mean_squared_error(predict_y, y_test))} \nThe R2 score is {r2_score(predict_y, y_test)}")
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic = ElasticNet()
```

```python
elastic.fit(x_train, y_train)
```

```python
predict_y = elastic.predict(x_test)
```

```python
print(f"Metrics for {elastic.__class__.__name__}:")
```

```python
print(f"The mean absolute error is {mean_absolute_error(predict_y, y_test)}\nThe mean squared error is {mean_squared_error(predict_y, y_test)}\nThe root mean squared error is {np.sqrt(mean_squared_error(predict_y, y_test))} \nThe R2 score is {r2_score(predict_y, y_test)}")
```

```python
print("The random forest regressor is the best model.")
```
